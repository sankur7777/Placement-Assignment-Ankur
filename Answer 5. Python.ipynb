```python
import requests
import json

def download_and_extract_data(url):
    # Download the data from the API link
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code != 200:
        print("Failed to download the data.")
        return
    
    # Parse the JSON response
    data = json.loads(response.text)
    
    # Extract the show information
    show_name = data['name']
    show_language = data['language']
    show_genres = data['genres']
    
    # Extract the episode information
    episodes = data['_embedded']['episodes']
    
    # Print the show information
    print(f"Show Name: {show_name}")
    print(f"Language: {show_language}")
    print(f"Genres: {', '.join(show_genres)}")
    
    # Print the episode information
    print("Episode Information:")
    for episode in episodes:
        episode_name = episode['name']
        episode_season = episode['season']
        episode_number = episode['number']
        print(f"Season {episode_season}, Episode {episode_number}: {episode_name}")

# Specify the API link
url = "http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes"

# Call the function to download the data and extract the information
download_and_extract_data(url)
```
