```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.utils import shuffle
```

```python
# Reduction of Data - dropping unnecessary columns
df = df.drop(['Administrative', 'Administrative_Duration', 'Informational', 'Informational_Duration', 
              'ProductRelated', 'ProductRelated_Duration'], axis=1)
```

```python
# Data Cleaning
# Hashtag/Mentions Removal
df['SpecialDay'] = df['SpecialDay'].apply(lambda x: re.sub(r'@[A-Za-z0-9]+','',str(x)))
df['Month'] = df['Month'].apply(lambda x: re.sub(r'@[A-Za-z0-9]+','',str(x)))
```

```python

# URL Removal
df['SpecialDay'] = df['SpecialDay'].apply(lambda x: re.sub(r'http\S+','',str(x)))
df['Month'] = df['Month'].apply(lambda x: re.sub(r'http\S+','',str(x)))
```

```python

# Stopwords Removal
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
df['Month'] = df['Month'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
```

```python

# Lowercase
df['Month'] = df['Month'].apply(lambda x: x.lower())
```

```python

# Stemming
porter = PorterStemmer()
df['Month'] = df['Month'].apply(lambda x: ' '.join([porter.stem(word) for word in x.split()]))
```

```python
# Removing Punctuations
df['SpecialDay'] = df['SpecialDay'].apply(lambda x: re.sub(r'[^\w\s]','',str(x)))
df['Month'] = df['Month'].apply(lambda x: re.sub(r'[^\w\s]','',str(x)))
```

```python

# Randomization
df = shuffle(df)
```

```python

# Final processed dataset
print("\nProcessed dataset:\n", df.head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
# Vectorization - TF-IDF
tfidf_vectorizer = TfidfVectorizer(use_idf=True)
tfidf_vector = tfidf_vectorizer.fit_transform(df['Month'])
print("\nTF-IDF Vector:\n", tfidf_vector)

# Vectorization - Bag of Words (BoW)
count_vectorizer = CountVectorizer()
count_vector = count_vectorizer.fit_transform(df['Month'])
print("\nBag of Words Vector:\n", count_vector)
```
